```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Supondo que df_brasil seja um DataFrame do pandas com colunas 'ano' e 'imigrantes'
fig,ax = plt.subplots(figsize=(8, 4))
ax.plot(df_brasil["ano"], df_brasil['imigrantes'])

# Define o localizador para os ticks principais do eixo x a cada 5 unidades
ax.xaxis.set_major_locator(MultipleLocator(5))

plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.title('Imigração ao Longo dos Anos')
plt.grid(True)
plt.show()
```

# Grafico de linha e boxplot em subplots

```python
fig,axs = plt.subplots(1,2,figsize=(15,5))


axs[0].plot(df_brasil["ano"],df_brasil['imigrantes'])
axs[0].set_title('imigração do brasil para o canada\n 1980 a 2013')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Numero de imigrantes')
axs[0].grid()


# subplot 2 - grafico boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('box-plot da imigração do brasil para o canada de 1980 a 02013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('numero de imigrantes')
axs[1].grid()
```

# comparativo dos maiores paises da america latina 

```python
import matplotlib.pyplot as plt

# criar um subplots
fig,axs = plt.subplots(2,2,figsize=(10,6))

axs[0,0].plot(df.loc["Brasil",anos])
axs[0,0].set_title("Brasil")
axs[0,0].grid(True, linestyle='--', linewidth=0.5, color='black')


# colonbia

axs[0,1].plot(df.loc["Colômbia",anos])
axs[0,1].set_title("Colômbia")
axs[0,1].grid(True, linestyle='--', linewidth=0.5, color='black')

#argentina

axs[1,0].plot(df.loc["Argentina",anos])
axs[1,0].set_title("Argentina")
axs[1,0].grid(True, linestyle='--', linewidth=0.5, color='black')

# peru 

axs[1,1].plot(df.loc["Peru",anos])
axs[1,1].set_title("Peru")
axs[1,1].grid(True, linestyle='--', linewidth=0.5, color='black')

fig.subplots_adjust(hspace=0.5,wspace=0.3)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat: # USO DE ARRAY BIDIMENSIONAL POR PADRÃO 
    ax.set_xlabel("Ano")
    ax.set_ylabel("Numero de imigrantes")

## DEFININDO LIMITES NO EIXO Y 
ymin = 0
ymax = 7000

for ax in axs.ravel():
    ax.set_ylim(ymin,ymax)

fig.suptitle("Análise de Imigração por País", fontsize=16)

fig.legend(loc='upper center', ncol=4, bbox_to_anchor=(0.5, -0.05), title='Legendas', fontsize=12)
```

```python
import matplotlib.pyplot as plt

# Criar os subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 6))

# Adicionar gráficos aos subplots com rótulos
axs[0, 0].plot(df.loc["Brasil", anos], label="Brasil",color="b")
axs[0, 0].set_title("Brasil")
axs[0, 0].grid(True, linestyle='--', linewidth=0.5, color='black')

axs[0, 1].plot(df.loc["Colômbia", anos], label="Colômbia",color="b")
axs[0, 1].set_title("Colômbia")
axs[0, 1].grid(True, linestyle='--', linewidth=0.5, color='black')

axs[1, 0].plot(df.loc["Argentina", anos], label="Argentina",color="b")
axs[1, 0].set_title("Argentina")
axs[1, 0].grid(True, linestyle='--', linewidth=0.5, color='black')

axs[1, 1].plot(df.loc["Peru", anos], label="Peru",color="b")
axs[1, 1].set_title("Peru")
axs[1, 1].grid(True, linestyle='--', linewidth=0.5, color='black')

# Ajustar o espaçamento entre os gráficos
fig.subplots_adjust(hspace=0.5, wspace=0.3)

# Configurar os ticks do eixo x
for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))
    ax.set_xlabel("Ano")
    ax.set_ylabel("Número de Imigrantes")

# Definir limites no eixo y
ymin = 0
ymax = 7000
for ax in axs.ravel():
    ax.set_ylim(ymin, ymax)

# Adicionar um título geral à figura
fig.suptitle("Análise de Imigração por País", fontsize=16)

# Adicionar uma legenda geral
# `bbox_to_anchor` é usado para posicionar a legenda fora dos subplots
# Exibir o gráfico
plt.show()
```

# preparar as cores 

```python
pais = america_sul_sorted["pais"]
```

```python
# Algoritmo de Percorrer e Colorir
cores = []

for pais in america_sul_sorted.index:
    if pais =='Brasil':
        cores.append("green")

    else:
        cores.append("silver")

fig,ax = plt.subplots(figsize=(18,5))
ax.barh(america_sul_sorted.index,america_sul_sorted['Total'],color=cores)
ax.set_title('imigração da America do Sul para o Canada')
ax.set_xlabel('numero de imigrantes')
ax.set_ylabel('nome dos paises',fontsize=14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)
plt.show()
```

```python
for i,v in enumerate (america_sul_sorted['Total']):
    print(f'o valor do indice: {i} valor:{v}')
```

```python
# Algoritmo de Percorrer e Colorir
cores = []

for pais in america_sul_sorted.index:
    if pais =='Brasil':
        cores.append("green")

    else:
        cores.append("silver")

fig,ax = plt.subplots(figsize=(18,5))
ax.barh(america_sul_sorted.index,america_sul_sorted['Total'],color=cores)
ax.set_title('imigração da America do Sul para o Canada')
ax.set_xlabel('numero de imigrantes')
ax.set_ylabel('nome dos paises',fontsize=14)
ax.xaxis.set_tick_params(labelsize = 12)
ax.yaxis.set_tick_params(labelsize = 12)


for i,v in enumerate (america_sul_sorted['Total']):
    #print(f'o valor do indice: {i} valor:{v}')
    ax.text(v,i,str(v),color='black',fontsize=12)
ax.set_frame_on(False)
ax.get_xaxis().set_visible(False)

plt.show()
```
