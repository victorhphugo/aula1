# Criando o gráfico de linhas

```python
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid()
#plt.scatter()
plt.show()
```

```python
# plt.plot(brasil)
```

```python
plt.figure(figsize=(8,4))
plt.plot(df_brasil['ano'],df_brasil['imigrantes'])
plt.title("imigração do brasil para o canada")
plt.xticks(['1980','1985','1990','1995','2000','2005','2010','2013'])
plt.yticks([500,1000,1500,2000,2500,3000])
plt.grid()
#plt.scatter()
plt.show()
```

# criando subplots

```python
fig,ax =plt.subplots(figsize=(8,4))
ax.plot(df_brasil["ano"],df_brasil['imigrantes'])
ax.set_title('imigração do brasil para o canada\n 1980 a 2013')
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
ax.set_xlabel('Ano')
ax.set_ylabel('Numero de imigrantes')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

# Supondo que df_brasil seja um DataFrame do pandas com colunas 'ano' e 'imigrantes'
fig,ax = plt.subplots(figsize=(8, 4))
ax.plot(df_brasil["ano"], df_brasil['imigrantes'])

# Define o localizador para os ticks principais do eixo x a cada 5 unidades
ax.xaxis.set_major_locator(MultipleLocator(5))

plt.xlabel('Ano')
plt.ylabel('Número de Imigrantes')
plt.title('Imigração ao Longo dos Anos')
plt.grid(True)
plt.show()
```

# Grafico de linha e boxplot em subplots

```python
fig,axs = plt.subplots(1,2,figsize=(15,5))


axs[0].plot(df_brasil["ano"],df_brasil['imigrantes'])
axs[0].set_title('imigração do brasil para o canada\n 1980 a 2013')
axs[0].xaxis.set_major_locator(plt.MultipleLocator(5))
axs[0].set_xlabel('Ano')
axs[0].set_ylabel('Numero de imigrantes')
axs[0].grid()


# subplot 2 - grafico boxplot
axs[1].boxplot(df_brasil['imigrantes'])
axs[1].set_title('box-plot da imigração do brasil para o canada de 1980 a 02013')
axs[1].set_xlabel('Brasil')
axs[1].set_ylabel('numero de imigrantes')
axs[1].grid()
```

# comparativo dos maiores paises da america latina 

```python
# criar um subplots
fig,axs = plt.subplots(2,2,figsize=(10,6))
axs[0,0].plot(df.loc["Brasil",anos])
axs[0,0].set_title("Brasil")

# colonbia

axs[0,1].plot(df.loc["Colômbia",anos])
axs[0,1].set_title("Colômbia")
#argentina

axs[1,0].plot(df.loc["Argentina",anos])
axs[1,0].set_title("Argentina")

# peru 

axs[1,1].plot(df.loc["Peru",anos])
axs[1,1].set_title("Peru")

fig.subplots_adjust(hspace=0.5,wspace=0.3)

for ax in axs.flat:
    ax.xaxis.set_major_locator(plt.MultipleLocator(5))

for ax in axs.flat:
    ax.set_xlabel("Ano")
    ax.set_ylabel("Numero de imigrantes")
```
